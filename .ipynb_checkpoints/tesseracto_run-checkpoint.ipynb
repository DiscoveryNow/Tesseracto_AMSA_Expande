```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import tesseracto_models as tsm
import pandas as pd
import numpy as np
import copy as cp
import time
import matplotlib.pyplot as plt
from pylab import rcParams
import statsmodels.api as sm
```

```python
##LEER DATA y creacion de variables auxiliares##
amsa = pd.read_csv('https://raw.githubusercontent.com/Carpae/Desafio-Expande/master/BD_Mantenimiento.csv')
amsa['FEEDERS_FUNCIONANDO'] = amsa[[
        'FEEDER_12_ON_OFF','FEEDER_14_ON_OFF',
        'FEEDER_16_ON_OFF','FEEDER_18_ON_OFF']].isin({'FUNCIONANDO'}).sum(1)

amsa["time"] = amsa["time"].astype('datetime64[ns]')

dataset = amsa[['time','VEL_FEEDER_220','VEL_FEEDER_225',
                'VEL_FEEDER_230','VEL_FEEDER_235','FEEDERS_FUNCIONANDO',
                'TORQUE_P5','P1.CV.005','P2.CV.005','CV.005']]

deltaT = np.array([(5) for i in range(len(dataset)-1)])
deltaT = np.concatenate((np.array([0]), deltaT))

dataset.insert(1, 'âˆ†t', deltaT)

limit = int(len(dataset)*0.8)

trainset = dataset.iloc[dataset.index < limit]
testset = dataset.iloc[dataset.index >= limit]

```

```python

AMSA = amsa.copy()
AMSA['date'] = AMSA['time'].astype(str).apply(lambda x : x[0:10])
AMSA['date'] = AMSA['date'].astype('datetime64[ns]')
AMSA_dia = AMSA.set_index(['date'], inplace = True)
AMSA_dia = AMSA.groupby(['date']).mean()

AMSA_dia['CV.005'].plot(figsize = (10, 5))

# =============================================================================
# Grafico de modelo aditivos
# =============================================================================
rcParams['figure.figsize'] = 24, 20
decompositionDia = sm.tsa.\
    seasonal_decompose(AMSA_dia[['CV.005']], model='additive')

fig = decompositionDia.plot()
plt.title('Modelo Aditivo')
plt.savefig('Aditivo_Dia.png')
plt.show()
```
